# Create a file india and Create a Directory world

```python
touch india
```

```python
touch world
```

# Check content permissions

```python
ls -l
```

# Change india file permissions - Full permissions to Owner(4+2+1=7), Read and write permissions to Group user(4+2=6), Execution permission to others(1)


```python
chmod 761 india
```

# Change World directory permissions - Full permission to owner(4+2+1=7), Read permission to Group user(4), execution permission to others(1)

```python
chmod 741 world
```

# Add a group called team101

```python
groupadd team101
```

# List all groups

```python
cat /etc/group
```

# Add a user called employee02

```python
adduser employee02
```

# List out all users

```python
cat /etc/passwd
```

# Change owner from root to employee02 for file - india

```python
chown employee02 india
```

# Change Group from root to team101 for file -india

```python
chgrp team101 india
```

# Create 2 users called tm1 and t2

```python
adduser tm1 && adduser tm2
```

# Add tm1 and tm2 users to team101 group

```python
gpasswd -a tm1 team101 && gpasswd -a tm2 team101 (or) gpasswd -M tm1,tm2 team101
```
